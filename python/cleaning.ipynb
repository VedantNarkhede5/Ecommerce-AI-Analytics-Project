```python
# ============================================================
#   E-Commerce Data Analysis Project
#   Step 2 — Data Cleaning
#   File : data_cleaning.py
#   Author: Your Name
# ============================================================

# --- Import Library ---
import pandas as pd                    # pandas is used for data analysis


# ============================================================
# STEP 1 — Load the Dataset
# ============================================================

df = pd.read_csv("train.csv")          # Load the raw CSV file into a DataFrame

print("=" * 50)
print("STEP 1 — Dataset Loaded")
print("=" * 50)
print(f"Total Rows    : {df.shape[0]}")   # Number of rows
print(f"Total Columns : {df.shape[1]}")   # Number of columns


# ============================================================
# STEP 2 — Show Basic Info
# ============================================================

print("\n" + "=" * 50)
print("STEP 2 — Basic Info")
print("=" * 50)
print(df.info())                           # Shows column names, data types, non-null counts


# ============================================================
# STEP 3 — Check Missing Values
# ============================================================

print("\n" + "=" * 50)
print("STEP 3 — Missing Values")
print("=" * 50)
print(df.isnull().sum())                   # Count of missing values in each column


# ============================================================
# STEP 4 — Fix Missing Values
# ============================================================

df['Postal Code'] = df['Postal Code'].fillna(0)    # Fill missing Postal Codes with 0
df['Postal Code'] = df['Postal Code'].astype(int)  # Convert from 42420.0 → 42420

print("\n" + "=" * 50)
print("STEP 4 — Missing Values Fixed")
print("=" * 50)
print(df.isnull().sum())                   # Should show 0 missing values now


# ============================================================
# STEP 5 — Remove Duplicate Rows
# ============================================================

before = df.shape[0]                       # Row count before removing duplicates
df = df.drop_duplicates()                  # Remove any duplicate rows
after = df.shape[0]                        # Row count after removing duplicates

print("\n" + "=" * 50)
print("STEP 5 — Duplicates Removed")
print("=" * 50)
print(f"Rows Before : {before}")
print(f"Rows After  : {after}")
print(f"Removed     : {before - after} duplicate rows")


# ============================================================
# STEP 6 — Fix Date Columns (Text → Real Date)
# ============================================================

# dayfirst=True because dates are in DD/MM/YYYY format
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Ship Date']  = pd.to_datetime(df['Ship Date'],  dayfirst=True)

print("\n" + "=" * 50)
print("STEP 6 — Date Columns Fixed")
print("=" * 50)
print(f"Order Date type : {df['Order Date'].dtype}")   # Should show datetime64
print(f"Ship Date type  : {df['Ship Date'].dtype}")    # Should show datetime64


# ============================================================
# STEP 7 — Standardize Text Columns
# ============================================================

# .str.strip()  → removes extra spaces before/after text
# .str.title()  → makes First Letter Capital in each word

df['Region']       = df['Region'].str.strip().str.title()
df['Segment']      = df['Segment'].str.strip().str.title()
df['Category']     = df['Category'].str.strip().str.title()
df['Ship Mode']    = df['Ship Mode'].str.strip().str.title()
df['Customer Name']= df['Customer Name'].str.strip().str.title()

print("\n" + "=" * 50)
print("STEP 7 — Text Columns Standardized")
print("=" * 50)
print(f"Regions    : {df['Region'].unique()}")
print(f"Segments   : {df['Segment'].unique()}")
print(f"Categories : {df['Category'].unique()}")


# ============================================================
# STEP 8 — Check Negative Sales Values
# ============================================================

print("\n" + "=" * 50)
print("STEP 8 — Checking Negative Sales")
print("=" * 50)
negative_sales = df[df['Sales'] < 0]          # Filter rows where Sales is negative
print(f"Negative Sales rows : {len(negative_sales)}")  # Should be 0

if len(negative_sales) > 0:
    print(negative_sales)                      # Show those rows if any exist
else:
    print("No negative sales found. Data is clean!")


# ============================================================
# STEP 9 — Add Useful New Columns
# ============================================================

# Days to Ship = Ship Date minus Order Date
df['Days to Ship'] = (df['Ship Date'] - df['Order Date']).dt.days

# Extract Year and Month from Order Date
df['Order Year']  = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month

print("\n" + "=" * 50)
print("STEP 9 — New Columns Added")
print("=" * 50)
print(f"New columns added : Days to Ship, Order Year, Order Month")
print(df[['Order Date', 'Ship Date', 'Days to Ship', 'Order Year', 'Order Month']].head(5))


# ============================================================
# STEP 10 — Final Check and Save Cleaned File
# ============================================================

print("\n" + "=" * 50)
print("STEP 10 — Final Summary")
print("=" * 50)
print(f"Total Rows    : {df.shape[0]}")
print(f"Total Columns : {df.shape[1]}")
print(f"Missing Values: {df.isnull().sum().sum()}")   # Total missing across all columns

# Save cleaned file
df.to_csv("cleaned_superstore.csv", index=False)

print("\n✅ Data Cleaning Completed Successfully!")
print("✅ Cleaned file saved as : cleaned_superstore.csv")
```
